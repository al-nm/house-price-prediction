```python
# Celda 1: Importaciones
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Celda 2: Cargar datos
df = pd.read_csv('data/raw/train.csv')
print(f"Shape del dataset: {df.shape}")
df.head()
```

```python
# Celda 3: Análisis de la variable objetivo
print(df['SalePrice'].describe())

fig, axes = plt.subplots(1, 2, figsize=(15,5))
sns.histplot(df['SalePrice'], kde=True, ax=axes[0])
axes[0].set_title('Distribución de SalePrice')

sns.histplot(np.log1p(df['SalePrice']), kde=True, ax=axes[1])
axes[1].set_title('Distribución de SalePrice (Log)')
plt.show()
```

```python
# Celda 5: Correlaciones
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()['SalePrice'].sort_values(ascending=False)
print("Top 15 features correlacionadas:\n", corr.head(15))

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df[corr.head(10).index].corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
