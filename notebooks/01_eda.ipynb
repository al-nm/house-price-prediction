```python
import streamlit as st
import pandas as pd
import joblib
import numpy as np
from src.features import feature_engineering

st.set_page_config(page_title="Predicción de Precios de Casas", layout="wide")
st.title("🏠 Predicción de Precios de Casas")

# Cargar modelo
@st.cache_resource
def load_model():
    return joblib.load("models/best_model.pkl")

model = load_model()

# Inputs del usuario
col1, col2 = st.columns(2)

with col1:
    overall_qual = st.slider("Calidad general (1-10)", 1, 10, 7)
    gr_liv_area = st.number_input("Área habitable (ft²)", 500, 5000, 1500)
    garage_cars = st.slider("Plazas de garage", 0, 4, 2)
    total_bsmt_sf = st.number_input("Área sótano (ft²)", 0, 3000, 800)

with col2:
    year_built = st.number_input("Año de construcción", 1900, 2025, 2000)
    full_bath = st.slider("Baños completos", 0, 4, 2)
    neighborhood = st.selectbox("Vecindario", ["NAmes", "CollgCr", "OldTown", "Edwards", "Somerst"])

# Botón de predicción
if st.button("Predecir Precio", type="primary"):
    input_data = pd.DataFrame({
        'OverallQual': [overall_qual],
        'GrLivArea': [gr_liv_area],
        'GarageCars': [garage_cars],
        'TotalBsmtSF': [total_bsmt_sf],
        'YearBuilt': [year_built],
        'FullBath': [full_bath],
        'Neighborhood': [neighborhood]
    })
    
    input_data = feature_engineering(input_data)
    prediction_log = model.predict(input_data)
    prediction = np.expm1(prediction_log[0])
    
    st.success(f"**Precio estimado: ${prediction:,.2f} USD**")
```
